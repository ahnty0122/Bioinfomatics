```python
import pandas as pd
import numpy as np
```

```python
sets = pd.read_csv("sets.csv")
sets
```

```python
sets['collection_id'].unique()
```

Pathway 출처 데이터베이스2

```python
sets_items = pd.read_csv("sets_items.csv")
sets_items
```

Pathway 출처 데이터베이스에서 homo sapiens 데이터만 빼오기

```python
sets_items_human = sets_items[sets_items['species_id'] == np.int(9606)]
sets_items_human
```

Protein-Protein interaction 데이터 (activation/inhibition)

```python
actions = pd.read_csv('actions_all.csv')
actions
```

Merge - Pathway 출처 데이터베이스끼리 merge

```python
merge_human_sets = pd.merge(sets_items_human, sets, left_on = 'set_id', right_on = 'set_id')
merge_human_sets
```

```python
merge_human_sets['species_id'].value_counts()
```

```python
comment = merge_human_sets[['set_id', 'item_id', 'preferred_name', 'comment', 'collection_id', 'title']]
comment
```

```python
comment['item_id'].value_counts()
```

Protein name annotation 데이터

```python
proteins = pd.read_csv('proteins.csv')
proteins
```

Protein name 데이터 - homo sapiens 데이터만 저장

```python
proteins_human = proteins[proteins['species_id']==np.int(9606)]
proteins_human
```

```python
names = proteins_human[['protein_id','preferred_name']]
names
```

Protein 파일로 actions(protein-protein interaction)데이터 프로틴 이름 annotation

```python
a = pd.merge(names, actions, left_on = 'protein_id', right_on = 'item_id_a', how = 'left')
a
```

```python
a.columns = ['protein_id', 'a_name', 'item_id_a', 'item_id_b', 'mode', 'action', 'is_directional', 'a_is_acting', 'score']
del a['protein_id']
a
```

```python
b = pd.merge(names, a, left_on = 'protein_id', right_on = 'item_id_b')
b
```

```python
b.columns = ['protein_id', 'b_name', 'a_name', 'item_id_a', 'item_id_b', 'mode', 'action', 'is_directional', 'a_is_acting', 'score']
del b['protein_id']
b
```

```python
b = b.reindex(columns = ['a_name', 'b_name', 'item_id_a', 'item_id_b', 'mode', 'action', 'is_directional', 'a_is_acting', 'score'])
b
```

```python
b.to_csv('actions_annotation.csv', index = False)
```

Pathway 출처 데이터베이스와 protein-protein actions 데이터 merge

```python
c = pd.merge(b, comment, left_on = 'a_name', right_on = 'preferred_name')
c
```

```python
c = c[['a_name', 'b_name', 'item_id_a', 'item_id_b', 'mode', 'action',
       'is_directional', 'a_is_acting', 'set_id', 'score','comment', 'collection_id', 'title']]
c.columns = ['a_name', 'b_name', 'item_id_a', 'item_id_b', 'mode', 'action',
       'is_directional', 'a_is_acting', 'set_id_a', 'score','comment_a', 'collection_id_a', 'title_a']
c
```

```python
c.to_csv('actions_join_a.csv', index = 'False')
```

```python
d = pd.merge(b, comment, left_on = 'b_name', right_on = 'preferred_name')
d
```

```python
d = d[['a_name', 'b_name', 'item_id_a', 'item_id_b', 'mode', 'action',
       'is_directional', 'a_is_acting', 'set_id', 'score','comment', 'collection_id', 'title']]
d.columns = ['a_name', 'b_name', 'item_id_a', 'item_id_b', 'mode', 'action',
       'is_directional', 'a_is_acting', 'set_id_b', 'score','comment_b', 'collection_id_b', 'title_b']
d
```

```python
d.to_csv('actions_join_b.csv', index = 'False')
```
