```python
import pandas as pd
```

```python
df = pd.read_csv('./cdk1_clue.txt', delimiter = '\t')
```

```python
df.columns = [0]

num = df[0].str.split('|').str[1]
num2 = []
for i in num:
    try:
        i = i.strip()
        num2.append(i)
    except:
        num2.append(i)
        continue
df['RANK'] = num2

drug = df[0].str.split('|').str[2]
drug2 = []
for i in drug:
    try:
        i = i.strip()
        drug2.append(i)
    except:
        drug2.append(i)
        continue
df['Drug Name'] = drug2

target = df[0].str.split('|').str[3]
target2 = []
for i in target:
    try:
        i = i.strip()
        target2.append(i)
    except:
        target2.append(i)
        continue
df['Target Name'] = target2

score = df[0].str.split('|').str[4]
score2 = []
for i in score:
    try:
        i = i.strip()
        score2.append(i)
    except:
        score2.append(i)
        continue
df['Binding Score'] = score2

del df[0]
```

```python
df = df.drop([df.index[0],df.index[1],df.index[6810]])
df
```

```python
df.to_csv('./cdk1_clue.txt', sep = '\t', index = False)
```
